```python
import numpy
import pandas
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
# load dataset
dataframe = pandas.read_csv("housing.csv", delim_whitespace=True, header=None)
dataset = dataframe.values
# split into input (X) and output (Y) variables
X = dataset[:,0:13] # remember last number is not included when given range u got 13 entries excluding 13th index
Y = dataset[:,13] #13th index is for answer or for truth values
```

```python
# define base model
from keras import models
from keras.layers import Dense
def baseline_model():
    model = models.Sequential()
    model.add(Dense(10,activation='relu',input_shape=(13,)))
    model.add(Dense(1))
    model.compile(optimizer='Adam',loss='mse',metrics=['mse'])
    return model
```

```python
# fix random seed for reproducibility
seed = 7
numpy.random.seed(seed)
# evaluate model with standardized dataset
estimator = KerasRegressor(build_fn=baseline_model, epochs=100, batch_size=5, verbose=0)
kfold = KFold(n_splits=10, random_state=seed)
results = cross_val_score(estimator, X, Y, cv=kfold)
print("Results: %.2f (%.2f) MSE" % (results.mean(), results.std()))
```

```python
# evaluate model with standardized dataset
numpy.random.seed(seed)
estimators = []
estimators.append(('standardize', StandardScaler()))
estimators.append(('mlp', KerasRegressor(build_fn=baseline_model, epochs=50, batch_size=5, verbose=0)))
pipeline = Pipeline(estimators)
kfold = KFold(n_splits=10, random_state=seed)
results = cross_val_score(pipeline, X, Y, cv=kfold)
print("Standardized: %.2f (%.2f) MSE" % (results.mean(), results.std()))
```

# Functional API

```python
import keras
def functional_api_model():
    inputs = keras.Input(shape=(13,))
    x=Dense(13,activation='selu')(inputs)
    x=Dense(6,activation='relu')(x)
    outputs =Dense(1)(x)
    model=keras.Model(inputs,outputs)
    model.compile(optimizer='Adam',loss='mse',metrics=['mse'])
    return model
```

```python
numpy.random.seed(seed)
estimators = []
estimators.append(('standardize', StandardScaler()))
estimators.append(('mlp', KerasRegressor(build_fn=functional_api_model, epochs=100, batch_size=5, verbose=0)))
pipeline = Pipeline(estimators)
kfold = KFold(n_splits=10, random_state=seed)
results = cross_val_score(pipeline, X, Y, cv=kfold)
print("functional_api: %.2f (%.2f) MSE" % (results.mean(), results.std()))
```

# Model SubClassing

```python
import keras
from keras.layers import Dense
class MyModel(keras.Model):
    
    def __init__(self):
        super(MyModel, self).__init__()
        self.dense1 = Dense(13,activation='selu')
        self.dense2 = Dense(6,activation='relu')
        self.dense3 = Dense(1)
        
    def call(self,inputs):
        x=self.dense1(inputs)
        x=self.dense2(x)
        return self.dense3(x)
    
def mymodel():    
    model = MyModel()
    model.compile(optimizer='Adam',loss='mse',metrics=['mse'])
    return model
```

```python
numpy.random.seed(seed)
estimators = []
estimators.append(('standardize', StandardScaler()))
estimators.append(('mlp', KerasRegressor(build_fn=mymodel, epochs=100, batch_size=5, verbose=0)))
pipeline = Pipeline(estimators)
kfold = KFold(n_splits=10, random_state=seed)
results = cross_val_score(pipeline, X, Y, cv=kfold)
print("mymodel: %.2f (%.2f) MSE" % (results.mean(), results.std()))
```

```python
import numpy as np
k = 10
num_val_samples = len(X) // k
num_epochs = 100
all_mae_histories = []

for i in range(k):
    print('processing fold #', i)
    val_data = X[i * num_val_samples: (i + 1) * num_val_samples]
    val_targets = Y[i * num_val_samples: (i + 1) * num_val_samples]
    partial_train_data = np.concatenate(
    [X[:i * num_val_samples],
    X[(i + 1) * num_val_samples:]],
    axis=0)
    partial_train_targets = np.concatenate(
    [Y[:i * num_val_samples],
    Y[(i + 1) * num_val_samples:]],
    axis=0)
    model = mymodel()
    history = model.fit(partial_train_data, partial_train_targets,
    validation_data=(val_data, val_targets),
    epochs=num_epochs, batch_size=5, verbose=1)
    mae_history = history.history
    all_mae_histories.append(mae_history)
```

```python
print("mean squared error= %.2f%%" % (np.mean(mae_history['mean_squared_error'])*0.1))
print("loss= %.2f%%" % (np.mean(mae_history['loss'])*0.1))
```
